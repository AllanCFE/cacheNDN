# Split lines

```python
def detect_addition_eviction(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    for line in lines:
        if line.count('ADDITION') >= 2 or line.count('EVICTION') >= 2 or (line.count('ADDITION') + line.count('EVICTION')) >= 2:
            if line.find('ADDITION') == -1 or line.find('EVICTION') == -1:
                type_to_split = 'ADDITION' if line.find('ADDITION') != -1 else 'EVICTION'

                first_index = line.find(type_to_split)
                second_index = line.find(type_to_split, first_index + 1)
                if second_index != -1:
                    line = line[:second_index] + '\n' + line[second_index:]
                
                new_lines.append(line)
            else:
                type_to_split = 'ADDITION' if line.find('ADDITION') > line.find('EVICTION') else 'EVICTION'

                lines= line.replace(type_to_split, '\n' + type_to_split, 1)
                new_lines.append(lines)
        else:
            new_lines.append(line)
    
    with open(file_path, 'w') as f:
        f.writelines(new_lines)
```

# Synth time

```python
def add_synth_time(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    for line in lines:
        partes = line.strip().split(',')

        if len(partes) == 3:
            # If the third part is not a number, or it has no value, we add a 0
            if partes[2] == '':
                # If there is a '\n', replace it with ',0\n'
                if line.find('\n') != -1:
                    line = line.replace('\n', ',0\n')
                else:
                    line += '0'
            
            new_lines.append(line)
        else:
            if line.find('ADDITION') == 0 or line.find('EVICTION') == 0:
                line = line.replace('\n', ',0\n')
                new_lines.append(line)
            else:
                new_lines.append(line)

    with open(file_path, 'w') as f:
        f.writelines(new_lines)
```

# Show errors

```python
def evict_calc (evictions_path):
    error_count = 0
    # Abra o arquivo de texto em modo de leitura
    with open(evictions_path, 'r') as evictions_file:
        for linha_num, linha in enumerate(evictions_file, start=1):
            # Divida cada linha em quatro partes usando a vírgula como separador
            partes = linha.strip().split(',')

            # Verifique se há exatamente quatro partes
            if len(partes) != 3:
                # Se a linha não tiver quatro partes, imprima um erro com o texto da linha
                # print(f'ERRO na linha {linha_num}: "{linha}"')
                error_count += 1

    return error_count
```

# Main 

```python
path = r"A:\UFSC\TCC\Analysis\TESTE.txt"

print('Errors before: ',evict_calc(path))
detect_addition_eviction(path)
add_synth_time(path)
print('Errors after: ',evict_calc(path))
```
